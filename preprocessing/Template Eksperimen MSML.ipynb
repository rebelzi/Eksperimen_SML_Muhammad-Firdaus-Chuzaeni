### **Sumber Dataset**

Dataset ini merupakan dataset diabetes yang populer digunakan untuk pembelajaran machine learning, khususnya untuk pemula yang ingin mempelajari teknik klasifikasi biner.

### **Deskripsi Dataset**
Dataset ini berisi data medis dari **768 pasien wanita keturunan India Pima** yang berusia minimal 21 tahun. Tujuan dari dataset ini adalah untuk **memprediksi apakah seorang pasien menderita diabetes atau tidak** berdasarkan beberapa pengukuran diagnostik.

### **Fitur-Fitur Dataset**
Dataset memiliki **9 kolom** dengan rincian sebagai berikut:

1. **Pregnancies**: Jumlah kehamilan
2. **Glucose**: Konsentrasi glukosa plasma (2 jam setelah tes toleransi glukosa oral)
3. **BloodPressure**: Tekanan darah diastolik (mm Hg)
4. **SkinThickness**: Ketebalan lipatan kulit trisep (mm)
5. **Insulin**: Insulin serum 2 jam (mu U/ml)
6. **BMI**: Body Mass Index (berat badan dalam kg/(tinggi dalam m)^2)
7. **DiabetesPedigreeFunction**: Fungsi silsilah diabetes (skor yang menunjukkan kemungkinan diabetes berdasarkan riwayat keluarga)
8. **Age**: Usia (tahun)
9. **Outcome**: Variabel target (0 = Tidak Diabetes, 1 = Diabetes)

### **Karakteristik Dataset**
- **Jumlah Data**: 768 records
- **Jumlah Fitur**: 8 fitur prediktif + 1 target variable
- **Tipe Data**: Numerik (continuous dan discrete)
- **Kelas Target**: Binary classification (0 dan 1)
- **Use Case**: Prediksi diagnosis diabetes mellitus

### **Tujuan Penggunaan**
Dataset ini sangat cocok untuk:
- Pembelajaran supervised machine learning (klasifikasi biner)
- Eksplorasi teknik preprocessing data medis
- Evaluasi berbagai algoritma machine learning untuk prediksi penyakit
- Studi kasus tentang ketidakseimbangan kelas (class imbalance)

# **5. Data Preprocessing**

Pada tahap ini, data preprocessing adalah langkah penting untuk memastikan kualitas data sebelum digunakan dalam model machine learning.

Jika Anda menggunakan data teks, data mentah sering kali mengandung nilai kosong, duplikasi, atau rentang nilai yang tidak konsisten, yang dapat memengaruhi kinerja model. Oleh karena itu, proses ini bertujuan untuk membersihkan dan mempersiapkan data agar analisis berjalan optimal.

Berikut adalah tahapan-tahapan yang bisa dilakukan, tetapi **tidak terbatas** pada:
1. Menghapus atau Menangani Data Kosong (Missing Values)
2. Menghapus Data Duplikat
3. Normalisasi atau Standarisasi Fitur
4. Deteksi dan Penanganan Outlier
5. Encoding Data Kategorikal
6. Binning (Pengelompokan Data)

Cukup sesuaikan dengan karakteristik data yang kamu gunakan yah. Khususnya ketika kami menggunakan data tidak terstruktur.

```python
print("=" * 60)
print("MEMULAI DATA PREPROCESSING")
print("=" * 60)

# Copy dataframe untuk preprocessing
df_processed = df.copy()

# 1. Handle Nilai 0 yang Tidak Wajar
print("\n[1] MENANGANI NILAI 0 YANG TIDAK WAJAR")
print("-" * 60)

# Kolom yang tidak boleh 0 (replace dengan NaN kemudian impute)
zero_not_allowed = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in zero_not_allowed:
    # Replace 0 dengan NaN
    df_processed[col] = df_processed[col].replace(0, np.nan)
    print(f"âœ“ {col}: {df_processed[col].isnull().sum()} nilai 0 diubah menjadi NaN")

# 2. Impute Missing Values
print("\n[2] IMPUTE MISSING VALUES DENGAN MEDIAN")
print("-" * 60)

for col in zero_not_allowed:
    median_value = df_processed[col].median()
    df_processed[col].fillna(median_value, inplace=True)
    print(f"âœ“ {col}: Missing values diisi dengan median = {median_value:.2f}")

print(f"\nTotal missing values setelah imputation: {df_processed.isnull().sum().sum()}")

# 3. Remove Duplicates
print("\n[3] MENGHAPUS DATA DUPLIKAT")
print("-" * 60)

before_dup = len(df_processed)
df_processed = df_processed.drop_duplicates()
after_dup = len(df_processed)
print(f"âœ“ Data duplikat dihapus: {before_dup - after_dup} rows")
print(f"âœ“ Shape setelah remove duplicates: {df_processed.shape}")

# 4. Handle Outliers menggunakan IQR method
print("\n[4] MENANGANI OUTLIERS (IQR METHOD)")
print("-" * 60)

def remove_outliers_iqr(df, columns):
    """Remove outliers using IQR method"""
    df_clean = df.copy()
    outliers_count = {}
    
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Count outliers
        outliers = ((df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)).sum()
        outliers_count[col] = outliers
        
        # Remove outliers
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    
    return df_clean, outliers_count

numeric_features = df_processed.columns[:-1].tolist()  # Semua kecuali Outcome
before_outliers = len(df_processed)
df_processed, outliers_info = remove_outliers_iqr(df_processed, numeric_features)
after_outliers = len(df_processed)

print(f"Outliers per kolom:")
for col, count in outliers_info.items():
    print(f"  â€¢ {col}: {count} outliers")

print(f"\nâœ“ Total rows dihapus: {before_outliers - after_outliers}")
print(f"âœ“ Shape setelah remove outliers: {df_processed.shape}")

# 5. Feature Scaling (Standardization)
print("\n[5] FEATURE SCALING (STANDARDIZATION)")
print("-" * 60)

# Separate features and target
X = df_processed.drop('Outcome', axis=1)
y = df_processed['Outcome']

# Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)

print("âœ“ Features di-standardize menggunakan StandardScaler")
print(f"\nStatistik setelah scaling (mean â‰ˆ 0, std â‰ˆ 1):")
print(X_scaled_df.describe().loc[['mean', 'std']])

# Combine back with target
df_final = X_scaled_df.copy()
df_final['Outcome'] = y.values

# 6. Train-Test Split
print("\n[6] TRAIN-TEST SPLIT")
print("-" * 60)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled_df, y, test_size=0.2, random_state=42, stratify=y
)

print(f"âœ“ Training set: {X_train.shape[0]} samples ({(X_train.shape[0]/len(df_final))*100:.1f}%)")
print(f"âœ“ Test set: {X_test.shape[0]} samples ({(X_test.shape[0]/len(df_final))*100:.1f}%)")
print(f"\nDistribusi Outcome di Training set:")
print(y_train.value_counts())
print(f"\nDistribusi Outcome di Test set:")
print(y_test.value_counts())

# 7. Visualisasi Perbandingan Before vs After Preprocessing
print("\n[7] VISUALISASI PERBANDINGAN")
print("-" * 60)

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

# Pilih beberapa fitur untuk visualisasi
viz_features = ['Glucose', 'BMI', 'Age']

for i, col in enumerate(viz_features):
    # Before preprocessing
    axes[i].hist(df[col], bins=30, alpha=0.7, color='red', edgecolor='black', label='Before')
    axes[i].set_title(f'{col} - Before Preprocessing', fontweight='bold')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
    axes[i].legend()
    axes[i].grid(alpha=0.3)
    
    # After preprocessing (scaled)
    axes[i+3].hist(X_scaled_df[col], bins=30, alpha=0.7, color='green', edgecolor='black', label='After')
    axes[i+3].set_title(f'{col} - After Preprocessing (Scaled)', fontweight='bold')
    axes[i+3].set_xlabel(col)
    axes[i+3].set_ylabel('Frequency')
    axes[i+3].legend()
    axes[i+3].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# 8. Save Preprocessed Data
print("\n[8] MENYIMPAN DATA YANG SUDAH DIPREPROCESS")
print("-" * 60)

# Create directory if not exists
import os
os.makedirs('diabetes_preprocessing', exist_ok=True)

# Save to CSV
df_final.to_csv('diabetes_preprocessing/preprocessed_diabetes.csv', index=False)
print("âœ“ Data disimpan ke: diabetes_preprocessing/preprocessed_diabetes.csv")

# Save train-test split
X_train.to_csv('diabetes_preprocessing/X_train.csv', index=False)
X_test.to_csv('diabetes_preprocessing/X_test.csv', index=False)
y_train.to_csv('diabetes_preprocessing/y_train.csv', index=False)
y_test.to_csv('diabetes_preprocessing/y_test.csv', index=False)
print("âœ“ Train-test split disimpan")

# Summary
print("\n" + "=" * 60)
print("RINGKASAN PREPROCESSING")
print("=" * 60)
print(f"Dataset awal: {df.shape}")
print(f"Dataset akhir: {df_final.shape}")
print(f"Rows dihapus: {df.shape[0] - df_final.shape[0]} ({((df.shape[0] - df_final.shape[0])/df.shape[0]*100):.1f}%)")
print(f"Features: {df_final.columns[:-1].tolist()}")
print(f"Target: Outcome")
print(f"\nData siap untuk training!")
print("=" * 60)

# Display sample preprocessed data
print("\nðŸ“Š Sample Data Setelah Preprocessing:")
display(df_final.head(10))
```
